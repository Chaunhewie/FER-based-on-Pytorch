# 单体模型检测

```python
# Testing
def test(lloader):
    global Test_acc
    net.eval()
    correct = 0
    total = 0
    cur_test_acc = 0.
    time_start = time.time()
    with torch.no_grad():
        for batch_idx, (inputs, targets) in enumerate(lloader):
            if transform_using_crop:
                bs, ncrops, c, h, w = np.shape(inputs)
                targets = torch.Tensor([[target]*ncrops for target in targets]).view(-1)
            else:
                bs, c, h, w = np.shape(inputs)
            inputs = inputs.view(-1, c, h, w)
            if use_cuda:
                inputs, targets = inputs.to(DEVICE), targets.to(DEVICE, torch.long)
            inputs, targets = Variable(inputs), Variable(targets)
            outputs = net(inputs)

            _, predicted = torch.max(outputs.data, 1)
            if target_type == 'ls':
                ground_value = targets.data
            elif target_type == 'fa':
                _, ground_value = torch.max(targets.data, 1)

            for i in range(len(predicted)):
                if predicted[i] == ground_value[i]:
                    c = predicted[i].item()
                    correct_map[c] += 1
                confuse_map[ground_value[i]][predicted[i]] += 1

            total += targets.size(0)
            correct += predicted.eq(ground_value.data).cpu().sum()
            cur_test_acc = float(correct) / float(total) * 100.

            time_end = time.time()
            duration = time_end - time_start
            utils.progress_bar(batch_idx, len(lloader), 'Time: %.2fs | Acc: %.3f%% (%d/%d)' %
                               (duration, cur_test_acc, correct, total))
```

# 双模型检测（本研究提出的）

```python
# Testing
def test(lloader, lloader_fl, weights_fl=0.5):
    global Test_acc
    net.eval()
    correct = 0
    total = 0
    cur_test_acc = 0.
    time_start = time.time()
    lloader_itr = enumerate(lloader)
    lloader_fl_itr = enumerate(lloader_fl)
    while True:
        try:
            bs, (inputs, targets) = next(lloader_itr)
            bs_fl, (inputs_fl, targets_fl) = next(lloader_fl_itr)
        except StopIteration:
            print('Over!')
            break
        if bs != bs_fl:
            assert 'Err: bs(%d) != bs_fl(%d)' % (bs, bs_fl)
        outputs = model_test(net, inputs, targets)
        outputs_fl = model_test(net_fl, inputs_fl, targets_fl)
        
        real_outputs = (1 - weights_fl) * outputs.data + weights_fl * outputs_fl.data
        _, predicted = torch.max(real_outputs, 1)
        if target_type == 'ls':
            ground_value = targets.data
        elif target_type == 'fa':
            _, ground_value = torch.max(targets.data, 1)

        for i in range(len(predicted)):
            if predicted[i] == ground_value[i]:
                c = predicted[i].item()
                correct_map[c] += 1
            confuse_map[ground_value[i]][predicted[i]] += 1

        total += targets.size(0)
        correct += predicted.eq(ground_value.data).cpu().sum()
        cur_test_acc = float(correct) / float(total) * 100.

        time_end = time.time()
        duration = time_end - time_start
        utils.progress_bar(bs, len(lloader), 'Time: %.2fs | Acc: %.3f%% (%d/%d)' %
                           (duration, cur_test_acc, correct, total))
        
def model_test(net, inputs, targets):
     with torch.no_grad():
        if transform_using_crop:
            bs, ncrops, c, h, w = np.shape(inputs)
            targets = torch.Tensor([[target]*ncrops for target in targets]).view(-1)
        else:
            bs, c, h, w = np.shape(inputs)
        inputs = inputs.view(-1, c, h, w)
        if use_cuda:
            inputs, targets = inputs.to(DEVICE), targets.to(DEVICE, torch.long)
        inputs, targets = Variable(inputs), Variable(targets)
        outputs = net(inputs)
        return outputs.cpu()
    
```
